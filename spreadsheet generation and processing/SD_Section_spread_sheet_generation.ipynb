### The code combined all the combined spreads sheet from present absent task and generated the sub-blocks spreadsheet for every block in the given task

```python
import glob
import os
import openpyxl
import numpy as np

###
block_num = 1 #counter count the block number
total_block = 4 #total number of block generated by the stimulus generator
#
root_dir = "E:/cmu/lab project/stimuli generation/syllable-4-semitone-interval-10/task_type_sd"
xlsx_files = []
for dirpath, dirnames, filenames in os.walk(root_dir):
    xlsx_files.extend(glob.glob(os.path.join(dirpath,'*.xlsx')))

xlsx_files = [i.replace('\\', '/') for i in xlsx_files]
xlsx_files = [name for name in xlsx_files if 'block' and 'combined'  in name]

print(len(xlsx_files)) # should be 24
print(xlsx_files[0])
```

```python
#Word
#pitchInstruction = '<p style="font-size: 32px;"> In this task, you will hear a sequence of words with different pitches, a pause, then another sequence with different pitches. Your task is to determine whether the PITCHes in the first sequence are in the same or different order than the PITCHes in the second sequence. After hearing these two sequences, you will be prompted to indicate whether these sequences are the "same" or "different". Press "Next" to start trials. </p>'
#wordInstruction = '<p style="font-size: 32px;"> In this task, you will hear a sequence of words with different pitches, a pause, then another sequence with different pitches. Your task is to determine whether the WORDs in the first sequence are in the same or different order than the WORDs in the second sequence. After hearing these two sequences, you will be prompted to indicate whether these sequences are the "same" or "different". Press "Next" to start trials. </p>'
#Syllable
pitchInstruction = '<p style="font-size: 32px;"> Remeber to attend PITCHes. Press "Next" to start trials. </p>'
wordInstruction = '<p style="font-size: 32px;"> Remeber to attend SYLLABLEs. Press "Next" to start trials. </p>'
#

blockIntro = '<p style="font-size: 32px; line-height: 1.2;">In this experiment there will be 4 blocks. <br> In this task, you will hear a sequence of syllables with different pitches, a pause, then another sequence with different pitches. <br>Your task is to determine whether the SYLLABLEs/PITCHes in the first sequence are in the same or different order than the SYLLABLEs/PITCHes in the second sequence. <br>After hearing these two sequences, you will be prompted to indicate whether these sequences are the "same" or "different". <br> Press "Next" to enter the block.  </p>'

output_dir  = "E:/cmu/lab project/stimuli generation/Task_SD/syllable-4semitone-10ms/SD_4_section_task"


for num in range (1, total_block+1):
    #create a new list of dir with only that block number
    block_name = 'Block_' + str(num)
    #print(block_name)
    sub_block = [name for name in xlsx_files if block_name  in name]
    #print(len(sub_block))
    
    #create a new xlsx file under the output_dir with name block _ num
    old_file1 = sub_block[0]
    old_wb1 = openpyxl.load_workbook(old_file1)
    old_ws1 = old_wb1.active

    #add the same first row
    new_wb = openpyxl.Workbook()
    new_ws = new_wb.active
    for i in range(1, old_ws1.max_column + 1):
        new_ws.cell(row=1, column=i).value = old_ws1.cell(row=1, column=i).value

    #append the section 
    new_ws.append(['','','text','','','',blockIntro])

    for sub_block_num in range(0, len(sub_block)):
        tp_ws = openpyxl.load_workbook(sub_block[sub_block_num]).active

        #add the 1st instruction
        if sub_block_num < 3:
            new_ws.append([sub_block_num + 1,'','text','','','',pitchInstruction,'','','','','','','',''])
        else: 
            new_ws.append([sub_block_num + 1,'','text','','','',wordInstruction,'','','','','','','',''])

        #add the xlsx from list all columns to the end of the new_ws       
        cur_ws_length = new_ws.max_row
        for i in range(2, tp_ws.max_row + 1):
            for j in range(2, tp_ws.max_column + 1):
                new_ws.cell(row=cur_ws_length+i-1, column=j).value = tp_ws.cell(row=i, column=j).value
            new_ws.cell(row=cur_ws_length+i-1, column=1).value = sub_block_num + 1
            new_ws.cell(row=cur_ws_length+i-1, column=2).value = sub_block_num + 1
        #change the randomized_blocks to num (optional)

    #create the new xlsx file
    new_file = output_dir + '/SD_section_' + str(num) + '.xlsx'
    new_wb.save(new_file)
```
